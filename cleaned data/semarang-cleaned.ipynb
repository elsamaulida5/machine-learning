### Library

```python
import numpy as np
import pandas as pd
```

### Gathering Data

```python
url = 'https://raw.githubusercontent.com/elsamaulida5/machine-learning/main/data/Laundry%20Semarang.csv'
data = pd.read_csv(url, encoding='latin1', na_values='*')
data
```

### Processing Data

#### Dropping Unnecessary Columns

```python
print(data.columns)
```

```python
columns_to_drop = ['Description', 'Street', 'Amenities', 'Phone', 'Cid', 'Fid', 'Place Id']
```

```python
data.drop(columns_to_drop, axis='columns', inplace=True)
```

```python
print(data.columns)
```

#### Deleting Missing Value

```python
data.isnull().sum()
```

For preventing the decreasing number of data, all the missing value will be substitute with "None".

```python
rows_to_change = ['Phones', 'Website', 'Domain', 'Opening Hours', 'Featured Image']
data[rows_to_change] = data[rows_to_change].fillna(value='None')
data
```

```python
data.isna().sum()
```

#### Checking for Duplicated Data

```python
duplicates = data.duplicated().sum()
print('The total of duplicated data is:', duplicates)
```

```python
duplicate_column = {}
for column in data.columns:
    duplicated_count = data[column].duplicated().sum()
    if duplicated_count > 0:
        duplicate_column[column] = duplicated_count

for column, count in duplicate_column.items():
    print(f"Column '{column} has {count} duplicated values")
```

```python
data = data.drop_duplicates(subset='Fulladdress')
```

```python
duplicate_column = {}
for column in data.columns:
    duplicated_count = data[column].duplicated().sum()
    if duplicated_count >= 0:
        duplicate_column[column] = duplicated_count

for column, count in duplicate_column.items():
    print(f"Column '{column} has {count} duplicated values")
```

data cleaned.

```python
output_name = 'semarang_cleaned.csv'
data.to_csv(output_name, index=False)

print(f"Cleaned data save to '{output_name}'")
```
