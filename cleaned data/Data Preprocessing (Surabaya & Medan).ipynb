# Data Preprocessing

```python
import pandas as pd
import numpy as np
from zipfile import ZipFile
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

## Data Laundry Surabaya

```python
df_surabaya = pd.read_csv("https://raw.githubusercontent.com/elsamaulida5/machine-learning/main/data/Laundry%20Surabaya.csv")
df_surabaya
```

```python
df_surabaya.info()
```

```python
# Checking Missing Values
df_surabaya.isnull().sum()
```

```python
# Because the "Amenities" and "Description" columns all data contain empty or null
# so the column will be dropped from the table

df_surabaya = df_surabaya.drop(columns=['Amenities','Description'])
```

```python
# Checking duplicate data
df_surabaya.duplicated(subset = ['Fulladdress']).sum()
```

```python
df_surabaya[df_surabaya.duplicated(subset=['Fulladdress'], keep=False)]
```

```python
df_surabaya = df_surabaya.drop_duplicates(subset = ['Fulladdress'])
```

```python
# Checking duplicate data
df_surabaya.duplicated(subset = ['Fulladdress']).sum()
```

```python
df_surabaya.isnull().sum()
```

```python
df_surabaya.info()
```

```python
# Drop the phone field because there is a phones field that contains phone numbers with country codes
df_surabaya = df_surabaya.drop(columns=['Phone', 'Street'])
```

```python
# Separates the two objects in the 'Phones' column
df_surabaya[['telephone','phone']] = df_surabaya['Phones'].str.split(', ', expand=True)
df_surabaya.drop(columns=['Phones'], inplace=True)
```

```python
df_surabaya.columns
```

```python
df_surabaya[['phone','telephone']]
```

```python
# Drop the telephone field
df_surabaya.drop(columns=['telephone'], inplace=True)

# Drop cid and fid columns
df_surabaya.drop(columns = ['Cid', 'Fid'], inplace = True)

# Drop Municipality field
df_surabaya.drop(columns = ['Municipality'], inplace = True)
```

```python
# Fill in the blanks with 'None' because the data in those columns might be needed
# so the blanks are replaced with 'None' to avoid missing values.

df_surabaya = df_surabaya.fillna('None')
df_surabaya
```

```python
df_surabaya.columns
```

```python
# Change the data value * to 0 
df_surabaya = df_surabaya.replace('*', 0)
```

```python
# Convert data types 'Review Count' and 'Average Rating' to float data type
df_surabaya['Average Rating'] = df_surabaya['Average Rating'].astype(float)
df_surabaya['Review Count'] = df_surabaya['Review Count'].astype(int)
```

```python
df_surabaya.info()
```
