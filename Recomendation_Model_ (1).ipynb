```python
#Membuat list uniqe dari data categories
def remove_duplicates(categories):
    split_categories = categories.split(',')
    unique_categories = set(cat.strip() for cat in split_categories)
    return ', '.join(unique_categories)

df['Unique_Categories'] = df['Categories'].apply(remove_duplicates)

unique_categories_set = set()
df['Unique_Categories'].str.split(', ').apply(unique_categories_set.update)

unique_categories_list = list(unique_categories_set)
unique_categories_list
```

```python
# Membuat keywords
# Fungsi untuk menetapkan keywords
def assign_keywords(services):
    keywords = []
    if 'Binatu' in services:
        keywords.extend(['just wash', 'iron wash'])
    if 'Binatu Cucian Kering' in services:
        keywords.append('dry cleaning')
    if 'Layanan Binatu' in services:
        keywords.extend(['just wash', 'iron wash', 'dry cleaning'])
    if 'koin' in services.lower():
        keywords.append('self laundry')
    if 'karpet' in services.lower():
       keywords.append('carpet laundry')
    if 'sepatu' in services.lower():
       keywords.append('shoe laundry')
    if 'jok' in services.lower():
       keywords.append('car seat laundry')
    return ', '.join(set(keywords))  # Menggunakan set untuk menghindari duplikasi keyword

# Terapkan fungsi ke DataFrame
df['Layanan'] = df['Categories'].apply(assign_keywords)

df
```

```python
# Menambahkan kolom ID Laundry ke kolom pertama
df.insert(0, 'ID Laundry', range(1, len(df) + 1))
df
```

# Membuat Model

```python
# Gabungkan deskripsi menjadi satu kolom
df['description'] = df['Fulladdress'] + ' ' + df['Categories'] + ' ' + df['Unique_Categories']
df['description']
```

```python
# Load pre-trained embedding model dari TensorFlow Hub
import tensorflow_hub as hub
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
# Buat embeddings untuk deskripsi laundry
descriptions = df['description'].tolist()
description_embeddings = embed(descriptions).numpy()
```

```python
# Persiapkan label
# Misalkan semua pencarian pengguna relevan dengan data laundry
# Oleh karena itu, label untuk semua data adalah 1.
labels = [1] * len(descriptions)
```

```python
# Model sederhana dengan TensorFlow
import tensorflow as tf
import numpy as np

model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(description_embeddings.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Kompilasi model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Latih model
model.fit(description_embeddings, np.array(labels), epochs=10)
```
