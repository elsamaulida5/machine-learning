```python
#Membuat list uniqe dari data categories
def remove_duplicates(categories):
    split_categories = categories.split(',')
    unique_categories = set(cat.strip() for cat in split_categories)
    return ', '.join(unique_categories)

df['Unique_Categories'] = df['Categories'].apply(remove_duplicates)

unique_categories_set = set()
df['Unique_Categories'].str.split(', ').apply(unique_categories_set.update)

unique_categories_list = list(unique_categories_set)
unique_categories_list
```

```python
# Membuat keywords
# Fungsi untuk menetapkan keywords
def assign_keywords(services):
    keywords = []
    if 'Binatu' in services:
        keywords.extend(['just wash', 'iron wash'])
    if 'Binatu Cucian Kering' in services:
        keywords.append('dry cleaning')
    if 'Layanan Binatu' in services:
        keywords.extend(['just wash', 'iron wash', 'dry cleaning'])
    if 'koin' in services.lower():
        keywords.append('self laundry')
    if 'karpet' in services.lower():
       keywords.append('carpet laundry')
    if 'sepatu' in services.lower():
       keywords.append('shoe laundry')
    if 'jok' in services.lower():
       keywords.append('car seat laundry')
    return ', '.join(set(keywords))  # Menggunakan set untuk menghindari duplikasi keyword

# Terapkan fungsi ke DataFrame
df['Layanan'] = df['Categories'].apply(assign_keywords)

df
```

```python
# Menambahkan kolom ID Laundry ke kolom pertama
df.insert(0, 'ID Laundry', range(1, len(df) + 1))
df
```

```python
laundry = df.drop(columns=['Unique_Categories'])
laundry.to_csv('laundry.csv', index=False)
laundry.info()
```

```python
# Gabungkan deskripsi menjadi satu kolom
laundry['description'] = laundry['Fulladdress'] + ' ' + laundry['Categories'] + ' ' + laundry['Layanan']
laundry['description']
```

```python
# Load pre-trained embedding model dari TensorFlow Hub
import tensorflow_hub as hub
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
# Buat embeddings untuk deskripsi laundry
descriptions = laundry['description'].tolist()
description_embeddings = embed(descriptions).numpy()
```

```python
# Membuat label target
import numpy as np
target = np.random.randint(0, 2, len(descriptions))
```

```python
# Membagi data menjadi pelatihan dan pengujian
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(description_embeddings, target, test_size=0.2, random_state=42)
```

```python
# Train Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Membangun model
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Melatih model
model.fit(X_train, y_train, epochs=1000, validation_split=0.2)
```

```python
# Evaluasi model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy}')
```

```python
def search_laundry(keyword, model, laundry, embed, threshold=0.1, max_results=30):
    # Buat embedding untuk keyword
    keyword_embedding = embed([keyword]).numpy()

    # Hitung embedding untuk deskripsi laundry
    descriptions = laundry['description'].tolist()
    description_embeddings = embed(descriptions).numpy()

    # Hitung kesamaan menggunakan dot product
    scores = np.dot(description_embeddings, keyword_embedding.T).flatten()

    # Tambahkan skor kecocokan ke dataframe laundry
    laundry['score'] = scores

    # Filter hasil dengan skor di atas threshold
    sorted_laundry = laundry.query('score > @threshold')

    # Urutkan laundry berdasarkan skor kecocokan
    sorted_laundry = sorted_laundry.sort_values(by='score', ascending=False)

    # Batasi jumlah hasil yang ditampilkan
    sorted_laundry = sorted_laundry.head(max_results)

    return sorted_laundry

# Contoh penggunaan pencarian dengan membatasi hasil menjadi 30 dan skor di atas 0.1
keyword = "dry cleaning"
result = search_laundry(keyword, model, laundry, embed, threshold=0.1)

# Tampilkan hasil
for index, row in result.iterrows():
    print(f"Nama Laundry: {row['Name']}")
    print(f"Alamat Lengkap: {row['Fulladdress']}")
    print(f"Telepon: {row['phone']}")
    print(f"Rating: {row['Average Rating']}")
    print(f"Website: {row['Website']}")
    print(f"Jam Buka: {row['Opening Hours']}")
    print(f"Layanan: {row['Layanan']}")
    print(f"score: {row['score']}")
    print("\n")
```
